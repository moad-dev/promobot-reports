```python
import pandas as pd
import re
```

```python
df = pd.read_csv('train_dataset_train.csv', sep=';')
```

```python
df
```

```python
text_col = df['Текст инцидента']
text_col
```

```python
#find in text_col texts any phone number and append it to list
phone_numbers = []
for text in text_col:
    regex = re.compile("\+?\d[\( -]?\d{3}[\) -]?\d{3}[ -]?\d{2}[ -]?\d{2}|\d{2,3}[ -]\d{2}[ -]\d{2}")
    if re.findall(regex, text):
        phone_numbers.append(re.findall(regex, text))
print(len(phone_numbers))
phone_numbers
```
